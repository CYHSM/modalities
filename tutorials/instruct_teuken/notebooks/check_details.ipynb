```python
from transformers import AutoTokenizer

# Load the same tokenizer used in your evaluation
tokenizer = AutoTokenizer.from_pretrained("/raid/s3/opengptx/mfrey/3.73T-Tokens/checkpoints/aug15_tokfix/instruct/checkpoint-80000/")
```

```python
# Decode the tokens to text
for i, detail in enumerate(details):
    if detail['cont_tokens']:
        # cont_tokens is likely nested (batch_size, sequence)
        tokens = detail['cont_tokens'][0]  # First (and likely only) item in batch
        
        # If it's further nested, you might need [0][0]
        if isinstance(tokens[0], list):
            tokens = tokens[0]
        
        # Decode tokens to text
        full_text = tokenizer.decode(tokens)
        
        print(f"Sample {i}:")
        print(f"Question: {detail['example']}")
        print(f"Gold answer: {detail['gold']}")
        print(f"Predicted (extracted): {detail['predictions']}")
        print(f"Full model output:")
        print(full_text)
        print("-" * 80)
```
